```python
import requests
from bs4 import BeautifulSoup
```

```python
# bahan
def get_bahan(resep_soup):
    try:
        li_bahan = resep_soup.find(class_='shortcode-ingredients').find('ul').find_all('li')
        str_bahan = ''
        for bahan in li_bahan:
            str_bahan += bahan.text + '\n'
        return str_bahan
    except:
        return None
```

```python
# resep
def get_resep(resep_soup):
    try:
        li_resep = resep_soup.find(class_='shortcode-directions').find('ol').find_all('li')
        str_resep = ''
        for langkah in li_resep:
            str_resep += langkah.text + '\n'
        return str_resep
    except:
        return None
```

```python
# judul
def get_judul(resep_soup):
    try:
        judul = resep_soup.find('h1').text
        return judul
    except:
        return None
```

```python
# url_gambar
def get_url_gambar(resep_soup):
    try:
        url_gambar = resep_soup.find(class_='post-thumb').find('img').attrs['src']
        return url_gambar
    except:
        return None
```

```python
link_resep = []
start = 1
end = 81
count = 0
for page in range(start, end):
    url = f'http://www.dapurkobe.co.id/semua/resep/page/{page}'
    hasil_request = requests.get(url)
    soup = BeautifulSoup(hasil_request.text, "html.parser")

    for hasil in  soup.find_all('div', class_='post'):
        count += 1
        print(count, hasil)
        link_resep.append(hasil.find('a').attrs['href'])
```

```python
count
```

```python
all_resep = []
count = 0
for link in link_resep:
    count += 1
    resep_soup = BeautifulSoup(requests.get(link).text, 'html.parser')
    resep_dict = {}
    resep_dict['url'] = link
    resep_dict['title'] = get_judul(resep_soup)
    resep_dict['image_url'] = get_url_gambar(resep_soup)
    resep_dict['bahan'] = get_bahan(resep_soup)
    resep_dict['resep'] = get_resep(resep_soup)
    print(count, resep_dict['url'])
    all_resep.append(resep_dict)
```

```python
import pandas as pd
df = pd.DataFrame(all_resep)
```
